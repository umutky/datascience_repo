```python
# Populasyon veri setimizde başlardan sonlara doğru aftertaste değeri düşmekte. Bunu göze almadan sampling yaparsak sample'ımızda bias ortaya çıkar.
# Bu yüzden verisetimizi shuffle'layalım.

shuffled = coffee.sample(frac=1)
shuffled = shuffled.reset_index(drop=True).reset_index()
shuffled.plot(x="index",y="aftertaste",kind="scatter")
plt.show()

# Gördüğümüz gibi artık bir düzene göre dağılmıyorlar.
# Bu verisetine şimdi systematic sampling yaparsak, basic sampling'den farkı kalmaz.
```

# Stratified Sampling

```python
coffee.head()
```

```python
coffee["country_of_origin"].value_counts(normalize=True).head(6)
```

```python
top_counted_countries = ['Mexico', 'Colombia', 'Guatemala', 'Brazil', 'Taiwan', 'United States (Hawaii)']
top_counted_subset = coffee["country_of_origin"].isin(top_counted_countries)
coffee_top = coffee[top_counted_subset]
```

```python
coffee_samp = coffee_top.sample(frac=0.1, random_state=1)

print("Sample")
print(coffee_samp["country_of_origin"].value_counts(normalize=True))
print("\nPopulation")
coffee_top["country_of_origin"].value_counts(normalize=True)
```

```python
# Proportional stratified sampling.
coffee_strat = coffee_top.groupby("country_of_origin").sample(frac=0.1, random_state=1)

coffee_strat["country_of_origin"].value_counts(normalize=True) # Very close the population.
```

```python
# Equal counts stratified sampling.
coffee_eq = coffee_top.groupby("country_of_origin").sample(n=15, random_state=1)

coffee_eq["country_of_origin"].value_counts(normalize=True)
```

# Weighted Random Sampling

taiwan'ın weight 2 yaptık diğerleri 1.

```python
import numpy as np
coffee_weight = coffee_top
condition = coffee_weight["country_of_origin"] == "Taiwan"
coffee_weight.loc[:,"weight"] = np.where(condition, 2, 1)
coffee_weight = coffee_weight.sample(frac=0.1, weights="weight")
```

```python
print("Population")
print(coffee_top["country_of_origin"].value_counts(normalize=True))
print("\n2x Weighted Taiwan")
print(coffee_weight["country_of_origin"].value_counts(normalize=True))
```

# Relative Error

```python
pop_mean = coffee["total_cup_points"].mean()
pop_mean
```

```python
sample_size = np.arange(5,1300, 1)
sample_mean = list()
for i in sample_size:
    sample_mean.append(coffee.sample(n=i)["total_cup_points"].mean())
```

```python
relative_error = []
for i in sample_mean:
    relative_error.append(100 * abs(pop_mean - i) / pop_mean)
```

```python
errors = pd.DataFrame({"sample_size" : sample_size,
                      "relative_error" : relative_error})
errors
```

```python
errors.plot(x="sample_size", y="relative_error", kind="line")
plt.xlim(0,1400)
plt.ylim(0,2.5)
plt.show()
```

**Genelde hesaplamalar yapılırken bütün popülasyona erişmek çok olası bir durum değildir, bu yüzden doğrudan sampling distribution yapmak çoğu zaman mümkün olmuyor. Bunun yerine Approximate Sampling Distribution kullanılır.**

https://pandas.pydata.org/pandas-docs/version/0.17.1/cookbook.html#creating-example-data

```python
import itertools
def expand_grid(data_dict):
    rows = itertools.product(*data_dict.values())
    return pd.DataFrame.from_records(rows, columns=data_dict.keys())
```

```python
# Expand a grid representing 5 8-sided dice
dice = expand_grid(
  {'die1': [1, 2, 3, 4, 5, 6, 7, 8],
   'die2': [1, 2, 3, 4, 5, 6, 7, 8],
   'die3': [1, 2, 3, 4, 5, 6, 7, 8],
   'die4': [1, 2, 3, 4, 5, 6, 7, 8],
   'die5': [1, 2, 3, 4, 5, 6, 7, 8]
  })

# Add a column of mean rolls and convert to a categorical
dice['mean_roll'] = (dice['die1'] + dice['die2'] + 
                     dice['die3'] + dice['die4'] + 
                     dice['die5']) / 5
dice['mean_roll'] = dice['mean_roll'].astype('category')

# Draw a bar plot of mean_roll
dice["mean_roll"].value_counts(sort=False).plot(kind="bar")
plt.show()
```

```python
# Sample one to eight, five times, with replacement
five_rolls = np.random.choice(list(range(1,9)), size=5, replace=True)

# Print the mean of five_rolls
print(five_rolls.mean())
```

# Bootstrapping

Sampling'in tam tersidir. Sample'dan teorik bir popülasyon elde etmeye çalışır.

```python
coffee_focus = coffee[["variety", "country_of_origin", "flavor"]]
coffee_focus = coffee_focus.reset_index()
coffee_focus.head()
```

```python
coffee_resamp = coffee_focus.sample(frac=1, replace=True)
```

```python
coffee_resamp.index.value_counts()
```

```python
num_unique_coffees = len(coffee_resamp.drop_duplicates(subset="index"))
num_unique_coffees
```

```python
# Coffee verisetinde olup coffee_resamp verisetinde olmayan kaç kahve var.
len(coffee) - num_unique_coffees
```

```python
coffee_samp = coffee[["variety", "country_of_origin", "flavor"]].reset_index().sample(n=500)
```

```python
mean_flavors_5000 = []
for i in range(5000):
    mean_flavors_5000.append(
        np.mean(coffee_samp.sample(frac=1, replace=True)["flavor"])
    )
bootstrap_distn = mean_flavors_5000
# it looks like a normal dist.
plt.hist(bootstrap_distn)
plt.show()
```

```python
# Mean, sample, boostrap and pop.
print("Mean of sample : ",coffee_samp["flavor"].mean())
print("Estimated mean with using bootstrap : ", np.mean(bootstrap_distn))
print("Population mean : ", coffee["flavor"].mean())
```

```python
# Std of SAMPLE, BOOTSTRAP and POP.
print("Std of sample : ",coffee_samp["flavor"].std())
print("Estimated std with using bootstrap : ", np.std(bootstrap_distn)) # çok az çıkıyor, çünkü aslında bu tahmin edilen std değildir.
print("Population mean : ", coffee["flavor"].std())
```

```python
# Standard error is the standard deviation of the statistic of interest.
standard_error = np.std(bootstrap_distn, ddof=1)
estimated_std = standard_error * np.sqrt(len(coffee_samp))
estimated_std
```

```python
# Calculate CI = 0.95 so alfa = 0.025
alfa=0.025

print("Mean of bootstrap dist",np.mean(bootstrap_distn))
print("Upper limit : ", np.quantile(bootstrap_distn, 1-alfa))
print("Lower limit : ", np.quantile(bootstrap_distn, alfa))
```

```python
# Generate a 95% confidence interval using the quantile method
lower_quant = np.quantile(bootstrap_distn, 0.025)
upper_quant = np.quantile(bootstrap_distn, 0.975)

# Print quantile method confidence interval
print((lower_quant, upper_quant))
```

```python
from scipy.stats import norm
# Find the mean and std dev of the bootstrap distribution
point_estimate = np.mean(bootstrap_distn)
standard_error = np.std(bootstrap_distn, ddof=1)

# Find the lower limit of the confidence interval
lower_se = norm.ppf(0.025, loc=point_estimate, scale=standard_error)

# Find the upper limit of the confidence interval
upper_se = norm.ppf(0.975, loc=point_estimate, scale=standard_error)

# Print standard error method confidence interval
print((lower_se, upper_se))
```
