```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/umutk/Desktop/BTK- VeriBilimi Bootcamp/Veri Setleri/Asım Egemen Yılmaz/dogrusal_regresyon_veriseti_20220209.csv', sep=';')
df.head()
```

```python
plt.scatter(df.deneyim, df.maas)
plt.xlabel('Deneyim yil')
plt.ylabel('Maas')
plt.title('Deneyim - Maas')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('maas', axis=1)
y = df['maas']

reg = LinearRegression().fit(X,y)
y_pred = reg.predict(X)
```

```python
reg.get_params()
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
r2_score(y, y_pred)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(X, y_pred, color='r')
plt.grid(True)
plt.show()
```

# Polinomial Regression

```python
df = pd.read_csv('C:/Users/umutk/Desktop/BTK- VeriBilimi Bootcamp/Veri Setleri/Asım Egemen Yılmaz/polinomsal_regresyon_veriseti_20220209.csv', sep=';')
df.head()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=df, x='araba_fiyat', y='araba_max_hiz');
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = df.drop('araba_max_hiz', axis=1)
y = df['araba_max_hiz']

X_polinom = PolynomialFeatures(degree=4).fit_transform(X)
X_polinom
```

```python
lr = LinearRegression()
lr.fit(X_polinom, y)
```

```python
y_pred = lr.predict(X_polinom)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.grid(True)
plt.show()
```

```python
r2_score(y, y_pred)
```
