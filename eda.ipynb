## Handling with Outliers

```python
sns.boxplot(data=df, y="Salary_USD")
plt.show()
```

```python
# IQR = 75th percentile - 25th percentile
seventy_fifth = df["Salary_USD"].quantile(0.75)
twenty_fifth = df["Salary_USD"].quantile(0.25)

IQR = seventy_fifth - twenty_fifth
print(f"IQR value is {IQR}")

upper_limit = seventy_fifth + 1.5*IQR
lower_limit = twenty_fifth - 1.5*IQR
print(f"upper limit is {upper_limit} and lower limit is {lower_limit}")

outlier_index = (df["Salary_USD"] < lower_limit)| (df["Salary_USD"] > upper_limit)
df[outlier_index]
```

```python
df_without_outliers = df[~outlier_index].reset_index(drop=True)
df_without_outliers

print("With Outliers")
sns.boxplot(data=df, y="Salary_USD")
plt.show()

print("Without Outliers")
sns.boxplot(data=df_without_outliers, y="Salary_USD")
plt.show()
```

```python
print("Before delete outliers")
sns.histplot(data=df, x="Salary_USD")
plt.show()
print("After delete outliers")
sns.histplot(data=df_without_outliers, x="Salary_USD")
plt.show()
```

# Relationships in Data

```python
divorce = pd.read_csv("/home/umut/python_calismalar/divorce.csv", parse_dates=["marriage_date", "divorce_date", "dob_man", "dob_woman"])
divorce.head()
```

```python
divorce.info()
```

**dt.month, dt.year, dt.day** : ile istediğimizi alabiliriz.

```python
divorce["marriage_month"] = divorce["marriage_date"].dt.month
divorce["marriage_month"]
```

```python
sns.lineplot(data=divorce, x="marriage_month", y="marriage_duration")
plt.show()
```

```python
# Define the marriage_year column
divorce["marriage_year"] = divorce["marriage_date"].dt.year

# Create a line plot showing the average number of kids by year
sns.lineplot(data=divorce, x=divorce["marriage_year"], y=divorce["num_kids"])
plt.show()
```

# Correlation

**.corr()**: pearson correlation kullanır ve bu yalnızca linear relation'ları değerlendirmekte iyidir. Non-linear correlation'ı düzgün değerlendiremiyor.

```python
cols = divorce.select_dtypes("number").columns
divorce[cols].corr()
```
