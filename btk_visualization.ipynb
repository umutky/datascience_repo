**'r+'** : red ve + seklinde olsun. 

```python
plt.plot(range(1,10), [x**2 for x in range(1,10)], 'r+')
plt.xlabel('tam sayilar')
plt.ylabel('tamsayilarin karesi')
plt.title('Bazi tamsayilarin kareleri')
plt.show()
```

```python
import numpy as np
t = np.arange(0.0, 5.0, 0.2)

plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^',
         markersize=3
        )
```

**Scatter Plot Example**

```python
veri = {'a' : np.arange(50),
        'c' : np.random.randint(0,50,50),
        'd' : np.random.randn(50)
       }

veri['b'] = veri['a'] + (10 * np.random.randn(50))
veri['d'] = np.abs(veri['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=veri)
plt.xlabel('a sutunundaki veriler')
plt.ylabel('b sutunundaki veriler')
plt.show()
```

**plt.hist(..., density=1)**

    In the context of a histogram in a plotting library like Matplotlib (assuming you're using Matplotlib based on the plt prefix), the density parameter is used to normalize the histogram. When density is set to 1, the histogram is normalized such that the area under the histogram is equal to 1. This means that the height of each bin is scaled in such a way that the total area of all the bins equals 1.

**r'....$\.....'**
    
    Bunu yaptigimizda ozel karakter gelecegini anliyor.


```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

plt.figure(figsize=(12,6))

n, bins, patches = plt.hist(x, bins=100, density=1, facecolor='b', alpha=0.40)
n, bins, patches = plt.hist(y, bins=100, density=1, facecolor='g', alpha=0.40)

plt.xlabel('Data')
plt.ylabel('Probabilities')
plt.title(r'Histogram of the data: $\mu_1=100,\ \mu_2=110$')
plt.annotate('Top of the Gauss Curve',
             xy=(110, 0.04),
             xytext=(130, 0.045),
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.text(75, 0.025, r'$\mu_1=100,\ \sigma_1=15$')
plt.text(120, 0.035, r'$\mu_1=110,\ \sigma_1=10$')
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 10, 0.01)
y = np.exp(x)

plt.figure()

plt.subplot(121)
plt.plot(x, y)
plt.yscale('linear')
plt.title('Lineer')
plt.grid(True)

plt.subplot(122)
plt.plot(x, y)
plt.yscale('log')
plt.title('Logarithm')
plt.grid(True)
```

# Scatterplot

```python
sns.set_style('darkgrid')
sns.scatterplot(data=data, x='boy', y='kilo', hue='cinsiyet', size='yas', sizes=(20,200), style='madalya', palette=['lightblue', 'pink'])
plt.show()
```
